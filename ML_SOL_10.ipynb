# 2. What is Base N encoding?
Base-N encoder encodes the categories into arrays of their base-N representation. For Binary encoding, the Base is 2 which means it converts the numerical values of a category into its respective Binary form.

# 3. Name suitable feature selection techniques. How you will do feature selection using Regularization?
Feature Selection is the method of reducing the input variable to your model by using only relevant data and getting rid of noise in data.
Some suitable Feature Selection techniques are:

*   Chi-square Test
*   Fisher's Score
*   Correlation Coefficient
*   Dispersion ratio
*   Backward Feature Elimination
*   Recursive Feature Elimination
*   Random Forest Importance
*   Entropy

Feature selction using Lasso and Ridge regularization.
**Lasso (L1)** regularization in many models causes many parameters to equal zero, so that the parameter vector is sparse.L1 is therefore useful for feature selection, as we can drop any variables associated with coefficients that go to zero
**Ridge (L2)** regularization doesnot shrink the parameters to 0. So it is not useful in feature selection.

# 4. What are different imputation techniques for handling categorical and numerical features?
Imputation for categorical features: One approach to imputing categorical features is to replace missing values with the most common class.
Imputation for Numerical features: One approach to imputing Numerical features is to replace missing values with the Mean/Median/Mode.

# 5. What is the difference between univariate and multivariate imputation?
Univariate imputation implies that only considering the values of a single column when performing imputation. Multivariate imputation, on the other hand, involves multiple features in the dataset when performing imputation.
